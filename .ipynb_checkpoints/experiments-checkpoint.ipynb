```python
import copy

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import numpy as np
import scipy.stats

import torch
from torch import nn
from torch.nn import functional as F
from torchvision import datasets, transforms, models
import tqdm

from ray import tune


import rgd
import function_benchmarks
import dl_benchmarks
```

Convex quadratics

```python
problem_size = 500
x0, obj_function, optimal_x, optimal_value, A, b = function_benchmarks.convex_quadratic(problem_size)
print(f'Objective function minimum: {optimal_value}')

iterations = 1000
```

Tune algo hyperparams on convex quadratic with Bayesian optimization

```python
quadratic_tune = function_benchmarks.tune_algos(
    x0,
    obj_function,
    iterations,
    tune_iters=20,
    hyperparam_space={
        "lr": tune.loguniform(1e-5, 5e-1),
        "momentum": tune.uniform(0.7, 1),
        "delta": tune.uniform(0, 40)
    },
    algos=["sgd", "cm", "nag", "adam", "rgd_eu", "rgd"]
)

print('Final hyperparam values:')
for algo, results in quadratic_tune.items():
    print(f'  {algo}: ')
    for hp, val in results['hyperparams'].items():
        print(f'    {hp}: {val}')
```

```python
results_df = quadratic_tune['cm']['analysis'].results_df

x = results_df['config.lr']
y = results_df['objective_value']
#plt.yscale('log')
plt.plot(x.sort_values(), y[x.argsort()], '.')
```

Run algos with our tuned hyperparams

```python
# SGD
sgd_quadratic_vals, sgd_quadratic_traj = function_benchmarks.run_optimizer(
    torch.optim.SGD, x0, obj_function, iterations, quadratic_tune['sgd']['hyperparams'])
print(f'SGD optimality gap: {sgd_quadratic_vals.min() - optimal_value}')

# Classical momentum
cm_quadratic_vals, cm_quadratic_traj = function_benchmarks.run_optimizer(
    torch.optim.SGD, x0, obj_function, iterations, quadratic_tune['cm']['hyperparams'])
print(f'CM optimality gap: {cm_quadratic_vals.min() - optimal_value}')

# Nesterov's accelerated gradient
nag_quadratic_vals, nag_quadratic_traj = function_benchmarks.run_optimizer(
    torch.optim.SGD, x0, obj_function, iterations, quadratic_tune['nag']['hyperparams'])
print(f'NAG optimality gap: {nag_quadratic_vals.min() - optimal_value}')

# Adam
adam_quadratic_vals, adam_quadratic_traj = function_benchmarks.run_optimizer(
    torch.optim.Adam, x0, obj_function, iterations, quadratic_tune['adam']['hyperparams'])
print(f'Adam optimality gap: {adam_quadratic_vals.min() - optimal_value}')

# Relativistic gradient descent with symplectic euler integrator
rgd_eu_quadratic_vals, rgd_eu_quadratic_traj = function_benchmarks.run_optimizer(
    rgd.RGD, x0, obj_function, iterations, quadratic_tune['rgd_eu']['hyperparams'])
print(f'RGD Euler optimality gap: {rgd_eu_quadratic_vals.min() - optimal_value}')

# Relativistic gradient descent with leapfrog integrator
rgd_quadratic_vals, rgd_quadratic_traj = function_benchmarks.run_optimizer(
    rgd.RGD, x0, obj_function, iterations, quadratic_tune['rgd']['hyperparams'])
print(f'RGD optimality gap: {rgd_quadratic_vals.min() - optimal_value}')
```

```python
plt.figure(figsize=(6, 4), dpi=150)

plt.plot(sgd_quadratic_vals - optimal_value, label='SGD')
plt.plot(cm_quadratic_vals - optimal_value, label='CM')
plt.plot(nag_quadratic_vals - optimal_value, label='NAG')
plt.plot(adam_quadratic_vals - optimal_value, label='Adam')
plt.plot(rgd_eu_quadratic_vals - optimal_value, label='RGD Euler')
plt.plot(rgd_quadratic_vals - optimal_value, label='RGD Leapfrog')

plt.yscale('log')
plt.legend()
plt.show()
```

Plotting trajectories on a 2-D slice of the parameter space

```python
all_trajs = np.concatenate([sgd_quadratic_traj[:, :2], cm_quadratic_traj[:, :2], nag_quadratic_traj[:, :2], adam_quadratic_traj[:, :2], rgd_eu_quadratic_traj[:, :2], rgd_quadratic_traj[:, :2]])

minimum = all_trajs.min() - np.abs(all_trajs.min()) * 0.2
maximum = all_trajs.max() + np.abs(all_trajs.max()) * 0.2
X = np.linspace(minimum, maximum)
Y = np.linspace(minimum, maximum)

Z = np.empty((len(Y), len(X)))
for i in range(len(X)):
    for j in range(len(Y)):
        x = torch.tensor(optimal_x.copy())
        x[0] = X[i]
        x[1] = Y[j]
        Z[j, i] = obj_function(x)

plt.figure(figsize=(10, 6), dpi=500)
plt.contourf(X, Y, Z, 30, cmap='RdGy')
plt.colorbar()

#plt.plot(sgd_quadratic_traj[:, 0], sgd_quadratic_traj[:, 1], label='SGD')
#plt.plot(cm_quadratic_traj[:, 0], cm_quadratic_traj[:, 1], label='CM')
plt.plot(nag_quadratic_traj[:, 0], nag_quadratic_traj[:, 1], label='NAG')
plt.plot(adam_quadratic_traj[:, 0], adam_quadratic_traj[:, 1], label='Adam')
#plt.plot(rgd_eu_quadratic_traj[:, 0], rgd_eu_quadratic_traj[:, 1], label='RGD Euler')
plt.plot(rgd_quadratic_traj[:, 0], rgd_quadratic_traj[:, 1], label='RGD Leapfrog')

#plt.plot(rgd_quadratic_traj[0, 0], rgd_quadratic_traj[0, 1], 'bo')
plt.plot(optimal_x[0], optimal_x[1], marker=(5,1), color='red', markersize=12)
plt.legend()

plt.plot()
plt.show()
```

```python
%load_ext tensorboard
%tensorboard --logdir tb_logs
```

```python
train_data = []
for k, batch in enumerate(train_loader):
    if k > 300: break
    train_data.append(batch)

test_data = []
for k, batch in enumerate(test_loader):
    if k > 300: break
    test_data.append(batch)
```

```python
model = copy.deepcopy(adam_model)

trajectory = np.array(model.trajectory)

width = trajectory[:, 0].max() - trajectory[:, 0].min()
height = trajectory[:, 1].max() - trajectory[:, 1].min()

X = np.linspace(trajectory[:, 0].min() - 0.5*width, trajectory[:, 0].max() + 0.5*width, num=20)
Y = np.linspace(trajectory[:, 1].min() - 0.5*height, trajectory[:, 1].max() + 0.5*height, num=20)
Z_train = np.empty((len(Y), len(X)))
Z_test = np.empty((len(Y), len(X)))

model.model.eval()
with torch.no_grad():

    for i in tqdm.trange(len(X)):
        for j in range(len(Y)):
            model.model[3].weight[0, 0] = X[i]
            model.model[3].weight[0, 1] = Y[j]

            train_losses = []
            for x, y in train_data:
                loss = model.loss_fn(model(x), y)
                train_losses.append(loss.item())

            test_losses = []
            for x, y in test_data:
                loss = model.loss_fn(model(x), y)
                test_losses.append(loss.item())

            Z_train[j, i] = np.mean(train_losses)
            Z_test[j, i] = np.mean(test_losses)
```

```python
plt.figure(figsize=(10, 6), dpi=500)
plt.contourf(X, Y, Z_train, 30, cmap='RdGy')
plt.colorbar()

plt.plot(np.array(adam_model.trajectory)[:, 0], np.array(adam_model.trajectory)[:, 1], label='Adam')
plt.plot(np.array(cm_model.trajectory)[:, 0], np.array(cm_model.trajectory)[:, 1], label='CM')
plt.plot(np.array(nag_model.trajectory)[:, 0], np.array(nag_model.trajectory)[:, 1], label='NAG')
plt.plot(np.array(rgd_model.trajectory)[:, 0], np.array(rgd_model.trajectory)[:, 1], label='RGD')
plt.plot(*trajectory[0], 'bo')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi=500)
plt.contourf(X, Y, Z_test, 30, cmap='RdGy')
plt.colorbar()

plt.plot(np.array(adam_model.trajectory)[:, 0], np.array(adam_model.trajectory)[:, 1])
plt.plot(np.array(cm_model.trajectory)[:, 0], np.array(cm_model.trajectory)[:, 1])
plt.plot(np.array(nag_model.trajectory)[:, 0], np.array(nag_model.trajectory)[:, 1])
plt.plot(np.array(rgd_model.trajectory)[:, 0], np.array(rgd_model.trajectory)[:, 1])
plt.plot(*trajectory[0], 'bo')

plt.plot()
plt.show()
```
